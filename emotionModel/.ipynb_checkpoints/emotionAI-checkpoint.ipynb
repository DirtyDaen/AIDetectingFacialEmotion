```python
import tensorflow as tf
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
Datadirectory = "Training/" #training dataset
```

```python
Classes = ["0", "1", "2", "3", "4", "5", "6"] #list of classes (exact folder name needed)
```

```python
for category in Classes:
    path = os.path.join(Datadirectory, category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img))
        plt.imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))
        plt.show()
        break
    break
```

```python
img_size = 224
new_array = cv2.resize(img_array, (img_size, img_size))
plt.imshow(cv2.cvtColor(new_array, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
new_array.shape
```

Read all images and convert into array

```python
training_Data = [] #data array

def create_training_Data():
    for category in Classes:
        path = os.path.join(Datadirectory, category)
        class_num = Classes.index(category) #label
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img))
                new_array = cv2.resize(img_array, (img_size, img_size))
                training_Data.append([new_array, class_num])
            except Exception as e:
                pass
```

```python
create_training_Data()
```

```python
print(len(training_Data))
```

```python
import random

random.shuffle(training_Data)
```

```python
X = [] #data
y = [] #label

for features, label in training_Data:
    X.append(features) #array with images
    y.append(label) #label
    
X = np.array(X).reshape(-1, img_size, img_size, 3) #converting into 4 dimensions
```

```python
X.shape
```

```python
X = X/255.0; #normalize data
```

```python
Y = np.array(y)
```

```python
Y.shape
```

deep learning model for training

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.MobileNetV2() #pre-trained model
```

```python
model.summary()
```

transfer learning

```python
base_input = model.layers[0].input #input
```

```python
base_output = model.layers[-2].output
```

```python
base_output
```

```python
final_output = layers.Dense(128)(base_output) #adding new layer, after output of global pooling layer
final_output = layers.Activation('relu')(final_output) #activation function
final_output = layers.Dense(64)(final_output) 
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(7, activation = 'softmax')(final_output) #7 because of 7 classes | classification layer
```

```python
final_output #output
```

```python
new_model = keras.Model(inputs = base_input, outputs = final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss = "sparse_categorical_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

```python
new_model.fit(X, Y, epochs = 2)
```

```python
new_model.save('my_model_01p07.h5')
```
